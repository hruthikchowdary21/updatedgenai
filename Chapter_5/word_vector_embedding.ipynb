```python
from gensim.models import Word2Vec

# Sample sentences
sentences = [
    ['the', 'king', 'loves', 'his', 'queen'],
    ['a', 'queen', 'rules', 'her', 'kingdom'],
    ['a', 'man', 'loves', 'his', 'country']
]

# Train a Word2Vec model
model = Word2Vec(sentences, vector_size=50, window=3, min_count=1, workers=4)

# Get the word vector for "king"
king_vector = model.wv['king']
queen_vector= model.wv['queen']
man_vector= model.wv['man']
print("Vector for 'king':", king_vector)
print("Vector for 'queen':", queen_vector)
print("Vector for 'man':", man_vector)
```
