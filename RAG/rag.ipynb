```python
import pandas as pd
from pymongo import MongoClient
from sentence_transformers import SentenceTransformer
import docx
```

```python
doc = docx.Document("C:/Users/hruth/GENAI/RAG/rrr.docx")
```

```python
# Extracting text chunks from the document
text_chunks = [para.text for para in doc.paragraphs if para.text]
```

```python
# Embedding the text chunks
model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(text_chunks)
```

```python
# Convert embeddings to lists and numpy floats to Python floats
df = pd.DataFrame({
    'text': text_chunks,
    'embedding': [list(map(float, embedding)) for embedding in embeddings]
})
```

```python
# Connecting to MongoDB
client = MongoClient("mongodb://localhost:27017/")
```

```python
# Creating a new database and collection
db = client['text_embeddings']
collection = db['embeddings']
```

```python
# Inserting the data into MongoDB
collection.insert_many(df.to_dict('records'))
```

```python
# Verifying the inserted data
for document in collection.find().limit(5):
    print(document)
```
